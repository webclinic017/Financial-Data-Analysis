```python
import datetime as dt
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from mpl_finance import candlestick_ohlc
import pickle
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

from bs4 import BeautifulSoup as bs
import requests
```

```python
html = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies').text
soup = bs(html, 'lxml')
soup
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
table
```

```python
tickers = []
rows = table.find_all('tr')[1:]
# rows
for row in rows:
    ticker = row.find_all('td')[0].text
    tickers.append(ticker[:-1])
    
tickers
```

```python
with open('sp500tickers.pickle', 'wb') as f:
    pickle.dump(tickers, f)
```

```python
with open('sp500tickers.pickle', 'rb') as f:
    tickers = pickle.load(f)
    print(tickers)
```

```python
# Create a csv file for all different tickers
for ticker in tqdm(tickers):
    ticker = ticker.replace(' ', '')
    try:
        df=pdr.get_data_yahoo(ticker, dt.datetime(2022, 1, 1), dt.datetime.now())
        df.to_csv('S&P500/{}.csv'.format(ticker))
        print(ticker)
    except:
        print('Error: {}'.format(ticker))
```

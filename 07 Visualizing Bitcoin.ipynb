## Visualizing Bitcoin in Python

```python
import datetime as dt
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import mplfinance as mpf
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

from bs4 import BeautifulSoup as bs
import requests
```

```python
crypto = "BTC"
currency = "USD"
```

```python
start = dt.datetime(2022, 1, 1) # start date
end = dt.datetime.now()
```

```python
df = pdr.DataReader(f"{crypto}-{currency}", 'yahoo', start, end)
df.head()
```

```python
mpf.plot(df, type='candle', volume=True, style='yahoo')
```

```python
cryptos = ["BTC","ETH","SOL"]
for crypto in cryptos:
    print(f"Prices of {crypto} for last {len(df)} days")
    mpf.plot(pdr.DataReader(f"{crypto}-{currency}", 'yahoo', start, end), type='candle', volume=True, style='yahoo')
```

```python
# Plotting all coins in one graph
cryptos = ["BTC","ETH","SOL"]

# Display the graphs in one plot 
for crypto in cryptos:
    df = pdr.DataReader(f"{crypto}-{currency}", 'yahoo', start, end)
    plt.plot(df.index, df['Adj Close'],'-', label=f"{crypto}")
    # Add legend
    plt.legend(loc='best')
    plt.xlabel('Date')
    plt.ylabel('Price USD')
    plt.yscale('log')
        
    plt.xlim(start, end)
    plt.xticks(rotation=45);
```

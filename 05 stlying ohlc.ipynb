## Updated Candlesticks

```python
import datetime as dt
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import mplfinance as mpf
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

from bs4 import BeautifulSoup as bs
import requests
```

```python
start = dt.datetime(2022, 1, 1) # start date is 2021-01-01
end = dt.datetime.now()

ticker = 'AAPL' # ticker is AAPL
df = pdr.get_data_yahoo(ticker, start, end) # get data from yahoo finance
df.head()
```

```python
mpf.available_styles()
```

```python
mpf.plot(df)
```

```python
for style in mpf.available_styles():
    print(f"Style - {style.capitalize()}")
    mpf.plot(df, style=style)
```

```python
# Creating custom styles
colors = mpf.make_marketcolors(up='g',down='r',wick='inherit',edge='inherit',volume='inherit')
type(colors)
colors
```

```python
mpf_style = mpf.make_mpf_style(base_mpf_style='yahoo',marketcolors=colors)
```

```python
mpf.plot(df, style=mpf_style)
```
